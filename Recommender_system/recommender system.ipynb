```python
path="page2full.json"
def load_items(path):
    with open(path, encoding="utf-8") as f:
        raw = json.load(f)

    # raw["data"] is a dict keyed by string IDs: {"100": {...}, ...}
    # Convert to list of item dicts
    if isinstance(raw, dict) and "data" in raw:
        data_block = raw["data"]
        if isinstance(data_block, dict):
            records = list(data_block.values())
        elif isinstance(data_block, list):
            records = data_block
        else:
            raise ValueError("Unexpected 'data' type")
    elif isinstance(raw, list):
        records = raw
    else:
        raise ValueError("Unknown JSON shape")

    # Flatten nested multilingual fields
    df = pd.json_normalize(records)  # e.g. name.en, name.ar, short_description.en, ...
    return df

df = load_items("page2full.json")  
df.head()
```

```python
NAME_COLS = [c for c in df.columns if c.startswith("name.")]
DESC_COLS = [c for c in df.columns if c.startswith("short_description.")]
TITLE_COLS = NAME_COLS  # alias

def join_nonempty(cells):
    parts = []
    for x in cells:
        if isinstance(x, str):
            s = x.strip()
            if s:
                parts.append(s)
    return " \n".join(parts)
def detect_langs(columns):
    langs = set()
    for c in columns:
        parts = c.split(".")
        if len(parts) >= 2 and parts[0] in {"name", "short_description", "description"}:
            langs.add(parts[1])
    return sorted(langs)

# df["text_multilang"] = [
#     join_nonempty([*(df.loc[i, TITLE_COLS].tolist()), *(df.loc[i, DESC_COLS].tolist())])
#     for i in range(len(df))
# ]

# LANGS = sorted(set([c.split(".")[1] for c in NAME_COLS]))
# for lang in LANGS:
#     t_name = df.get(f"name.{lang}", pd.Series([""]*len(df)))
#     t_desc = df.get(f"short_description.{lang}", pd.Series([""]*len(df)))
#     df[f"text_{lang}"] = [join_nonempty([t_name[i], t_desc[i]]) for i in range(len(df))]
LANGS = detect_langs(df.columns)        # e.g. ['ar','en','sq']

# Per-language text (collects ANY column that belongs to that language)
for lang in LANGS:
    name_cols  = [f"name.{lang}"] if f"name.{lang}" in df.columns else []
    short_cols = [f"short_description.{lang}"] if f"short_description.{lang}" in df.columns else []
    desc_cols  = [c for c in df.columns if c.startswith(f"description.{lang}")]
    cols = name_cols + short_cols + sorted(desc_cols)
    df[f"text_{lang}"] = df[cols].apply(lambda r: join_nonempty(r.dropna().tolist()), axis=1) if cols else ""

# Unified multilingual text
df["text_multilang"] = df[[f"text_{l}" for l in LANGS]].apply(
    lambda r: join_nonempty(r.dropna().tolist()), axis=1
)

# Persist items
df.to_parquet("items.parquet", index=False)
df.head()
```

```python
import torch
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer

# If CUDA not available, fallback to CPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Correct official Hugging Face repo path
MODEL_NAME = "entence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
model = SentenceTransformer(MODEL_NAME, device="cpu")  # Force CPU to avoid VRAM/memory issues


# model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
# embeddings = model.encode(sentences)
# print(embeddings)

# Define LANGS and join function
LANGS = ["ar", "en", "sq"]

def join_nonempty(cells):
    """Join non-empty text cells into one string."""
    parts = [str(x).strip() for x in cells if isinstance(x, str) and str(x).strip()]
    return " \n".join(parts)

def make_input_row(row):
    """Build a multilingual text block with title and description."""
    title = join_nonempty([row.get(f"name.{l}", "") for l in LANGS])
    desc  = join_nonempty([row.get(f"short_description.{l}", "") for l in LANGS] +
                        [row.get(f"text_{l}", "") for l in LANGS])
    return f"[TITLE]\n{title}\n[DESC]\n{desc}"

# Assuming df is your DataFrame
inputs = [make_input_row(df.iloc[i]) for i in range(len(df))]

# Encode in batches to avoid memory overflow
embs = model.encode(
    inputs,
    batch_size=32,  # Lower if still hitting memory issues
    show_progress_bar=True,
    convert_to_numpy=True,
    normalize_embeddings=True
)

# Save embeddings to file
np.save("item_embeddings.npy", embs)

print(f"Embeddings saved: {embs.shape}")
```
